```python
with plt.style.context('seaborn-white'):
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,7), sharex='col', sharey='row')
    axes[0,0].plot(retina_p_tsne['Perplexity'].astype(str), retina_p_tsne['R'], color='k', marker='o', ls='--', ms=10, linewidth=3)
    axes[1,0].plot(retina_p_tsne['Perplexity'].astype(str), retina_p_tsne['EMD'], color='k', marker='o', ls='--', ms=10, linewidth=3)
    axes[2,0].plot(retina_p_tsne['Perplexity'].astype(str), retina_p_tsne['Knn Pres.'], color='k', marker='o', ls='--', ms=10, linewidth=3)
    
    axes[0,1].plot(retina_p_umap['Perplexity'].astype(str), retina_p_umap['R'], color='k', marker='o', ls='--', ms=10, linewidth=3)
    axes[1,1].plot(retina_p_umap['Perplexity'].astype(str), retina_p_umap['EMD'], color='k', marker='o', ls='--', ms=10, linewidth=3)
    axes[2,1].plot(retina_p_umap['Perplexity'].astype(str), retina_p_umap['Knn Pres.'], color='k', marker='o', ls='--', ms=10, linewidth=3)
    
    axes[0,0].plot(colon_p_tsne['Perplexity'].astype(str), colon_p_tsne['R'], color='k', marker='o', ms=10, linewidth=3)
    axes[1,0].plot(colon_p_tsne['Perplexity'].astype(str), colon_p_tsne['EMD'], color='k', marker='o', ms=10, linewidth=3)
    axes[2,0].plot(colon_p_tsne['Perplexity'].astype(str), colon_p_tsne['Knn Pres.'], color='k', marker='o', ms=10, linewidth=3)
    
    axes[0,1].plot(colon_p_umap['Perplexity'].astype(str), colon_p_umap['R'], color='k', marker='o', ms=10, linewidth=3)
    axes[1,1].plot(colon_p_umap['Perplexity'].astype(str), colon_p_umap['EMD'], color='k', marker='o', ms=10, linewidth=3)
    axes[2,1].plot(colon_p_umap['Perplexity'].astype(str), colon_p_umap['Knn Pres.'], color='k', marker='o', ms=10, linewidth=3)
    
    axes[0,0].set_ylabel('R', fontsize='xx-large')
    axes[1,0].set_ylabel('EMD', fontsize='xx-large')
    axes[2,0].set_ylim([95, 100])
    axes[2,0].set_ylabel('Knn Pres. (%)', fontsize='xx-large')
    axes[2,0].set_xlabel('t-SNE Perplexity', fontsize='xx-large')
    axes[2,1].set_xlabel('UMAP n_neighbors', fontsize='xx-large')
    axes[0,0].tick_params(labelsize='xx-large')
    axes[1,0].tick_params(labelsize='xx-large')
    axes[2,0].tick_params(labelsize='xx-large')
    axes[2,1].tick_params(labelsize='xx-large')
    plt.legend(labels=['Retina','Colon'], fontsize='xx-large')
    plt.tight_layout()
    plt.savefig('archive/outputs/neighborhood_metrics.png', dpi=1000)
```
